```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
HR_PATH = "E:\\NCLOneDrive\\OneDrive - Newcastle University\\Dataset\\MESA\\TEST\\hr\\"
ACC_PATH = "E:\\NCLOneDrive\\OneDrive - Newcastle University\\Dataset\\MESA\\TEST\\act\\"
OVERLAP_PATH = "E:\\NCLOneDrive\OneDrive - Newcastle University\\Dataset\\MESA\\HR\\mesa\\polysomnography\\mesa-actigraphy-psg-overlap.csv"
OUTPUT_PATH = "E:\\NCLOneDrive\OneDrive - Newcastle University\\Dataset\\MESA\\Acc-HR-Aligned"
```

```python
def zero_leading_int(digits, length):
    assert isinstance(digits, int)
    to_fill = length - len(str(digits))
    if to_fill > 0:
        return str(digits).zfill(length)
    else:
        return str(digits)
```

```python
def process_all_files():
    all_acc_files = os.listdir(ACC_PATH)
    all_hr_files = os.listdir(HR_PATH)
    overlap_df = pd.read_csv(OVERLAP_PATH)

    for idx in np.arange(overlap_df.shape[0]):
        mesa_id = zero_leading_int(int(overlap_df['mesaid'].iloc[idx]), 4)
        try:
            acc_file_idx = all_acc_files.index(''.join([s for s in all_acc_files if mesa_id in s]))
            hr_file_idx = all_hr_files.index(''.join([s for s in all_hr_files if mesa_id in s]))
            acc_df = pd.read_csv(os.path.join(ACC_PATH, all_acc_files[acc_file_idx]))
            hr_df = pd.read_csv(os.path.join(HR_PATH, all_hr_files[hr_file_idx]))
            acc_start_idx = overlap_df['line'].iloc[idx].astype(int)
            hr_end_ms = int(hr_df['seconds'].tail(1))
            acc_epochs = int(np.floor(hr_end_ms / 30))
            acc_df = acc_df[acc_start_idx - 1: acc_start_idx + acc_epochs - 1]  
            # cut the data frame from the overlapped start index
            # !TODO calculate the RR intervals
            if (int(hr_df['epoch'].tail(1)) > acc_df.shape[0]):
                hr_df = hr_df[hr_df['epoch'] <= acc_df.shape[0]]
            for i in np.arange(acc_epochs):
                total = hr_df[hr_df['epoch'] == i + 1].shape[0]

            hr_df['RR Intervals'] = hr_df['seconds'].diff()
            hr_df['RR Intervals'].fillna(hr_df['RR Intervals'].mean(), inplace=True)
            # hr_df['RR Intervals'].apply(lambda x: hr_df['RR Intervals'].mean() if x == 0 else x)
            # hr_df['HR'] = hr_df['RR Intervals'].apply(lambda x: np.round((60.0/x)))
            hr_df['HR'] = np.round((60.0/hr_df['RR Intervals']),0)

        except Exception as e:
            print("ID: {}, Exception {}".format(mesa_id, e))
```

```python
process_all_files()
```
